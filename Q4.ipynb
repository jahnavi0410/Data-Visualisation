```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('crop_yield.csv')
selected_crops = ['Rice', 'Wheat', 'Maize', 'Cotton(lint)', 'Sugarcane']
area_data = data[data['Crop'].isin(selected_crops)].groupby('Crop')['Area'].sum()
plt.figure(figsize=(8, 8))
plt.pie(area_data, labels=area_data.index,startangle=90, colors=['pink', 'lightblue', 'orange', 'purple', 'green'])
plt.title('Area Distribution by Selected Crops')
plt.show()

```

```python

crop_data = data[data['Crop'] == 'Rice'].groupby('Crop_Year')['Yield'].mean()
plt.figure(figsize=(10, 6))
plt.plot(crop_data.index, crop_data.values, marker='o', linestyle='-', color='green')
plt.title('Rice Yield Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Yield (kg/ha)')
plt.grid(True)
plt.show()

```

```python
selected_crops = ['Rice', 'Wheat', 'Maize', 'Cotton(lint)', 'Sugarcane']
crop_production = data[data['Crop'].isin(selected_crops)].groupby('Crop')['Production'].sum()
plt.figure(figsize=(10, 6))
plt.bar(crop_production.index, crop_production.values)
plt.title('Total Production of Selected Crops')
plt.xlabel('Crop')
plt.ylabel('Total Production (in tons)', fontsize=12)
for index, value in enumerate(crop_production.values):
    plt.text(index, value + value * 0.05, f'{int(value):,}', ha='center')
plt.show()
```
