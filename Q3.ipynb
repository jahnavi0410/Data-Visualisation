```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
```

```python
sns.set(style="whitegrid")
```

```python
# Comparison Plot: Scatterplot for Sepal Length vs Sepal Width for different species
plt.figure(figsize=(10, 5))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Comparison of Sepal Dimensions across Species")
plt.show()
```

```python
# Composition Plot: Pie Chart for Species Distribution
species_counts = iris['species'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Species Composition in the Iris Dataset")
plt.show()
```

```python
# Comparison Plot: Boxplot for Petal Length across Species
plt.figure(figsize=(10, 5))
sns.boxplot(data=iris, x="species", y="petal_length")
plt.title("Petal Length Comparison across Species")
plt.show()
```

```python
# Composition Plot: Stacked Bar for Sepal Length Range Distribution
iris['sepal_length_range'] = pd.cut(iris['sepal_length'], bins=[4, 5, 6, 7, 8], labels=['4-5', '5-6', '6-7', '7-8'])
stacked_data = iris.groupby(['species', 'sepal_length_range']).size().unstack()
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title("Stacked Composition of Sepal Length Ranges by Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```
