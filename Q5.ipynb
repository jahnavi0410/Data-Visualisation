```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
plt.figure(figsize=(12, 10))
```

```python
# 1. Line Plot
plt.subplot(2, 3, 1)
plt.plot(tips['total_bill'], label='Total Bill', color='blue')
plt.xlabel("Index")
plt.ylabel("Total Bill")
plt.title("Line Plot of Total Bill")
plt.legend()
```

```python
# 2. Scatter Plot
plt.subplot(2, 3, 2)
plt.scatter(tips['total_bill'], tips['tip'], color='green')
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot of Tip vs. Total Bill")
```

```python
# 3. Bar Plot
plt.subplot(2, 3, 3)
day_counts = tips['day'].value_counts()
plt.bar(day_counts.index, day_counts.values, color='purple')
plt.xlabel("Day")
plt.ylabel("Count")
plt.title("Bar Plot of Day Counts")
```

```python
# 4. Histogram
plt.subplot(2, 3, 4)
plt.hist(tips['total_bill'], bins=10, color='orange')
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill")
```

```python
# 5. Box Plot
plt.subplot(2, 3, 5)
plt.boxplot([tips[tips['day'] == day]['total_bill'] for day in tips['day'].unique()], labels=tips['day'].unique())
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Box Plot of Total Bill by Day")
```

```python
# 6. Pie Chart
plt.subplot(2, 3, 6)
time_counts = tips['time'].value_counts()
plt.pie(time_counts, labels=time_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen'])
plt.title("Pie Chart of Time (Lunch/Dinner)")
```
